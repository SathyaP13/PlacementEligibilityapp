```python
# creating the dataset for placement eligibility app using OOPs concept
# Methodology used: Class, Methods and Objects

from faker import Faker
import pandas as pd
fake = Faker()

class dataset_creation:
# creating the dataset for students table and storing it in a excel file
# libraies used : Faker, Pandas
# datasets used from Python : List[], dictionary {}
# Methodology used : Functions, for loop
    def students(self,num_students):
        self.student_list=[]
        for i in range(1,num_students):
            student={}
            student['student_id']=i
            student['name']=fake.name()
            student['age']=fake.random_int(min=21,max=40)
            student['gender']=fake.random_element(elements=("Male","Female","Trans"))
            student['email']=fake.email()
            student['phone']=fake.phone_number()
            student['enrollment_year']=fake.random_int(2020,2025)
            student['course_batch']=fake.random_element(elements=("AIML-C-WD-E-B8","AIML-C-WE-A-C7","AIML-C-WE-A-C4","AIML-C-WD-E-B7"))
            student['city']=fake.city()
            student['graduation_year']=fake.random_int(2005,2020)
            self.student_list.append(student)
            df=pd.DataFrame(self.student_list)
            df= df.drop_duplicates(['student_id'],keep='first')
        return df

# creating the dataset for programming table and storing it in a excel file
# libraies used : Faker, Pandas
# datasets used from Python : List[], dictionary {}
# Methodology used : Functions, for loop

    def programming(self,prg):
        self.prog_list=[]
        for i in range(1,prg):
            prog={}
            prog['programming_id']=fake.unique.random_int(min=1,max=10000)
            prog['student_id']=i
            prog['language']=fake.random_element(elements=("Python","SQL","JAVA","JS"))
            prog['problems_solved']=fake.random_int(min=1,max=450,step=1)
            prog['assessments_completed']=fake.random_int(min=1,max=10,step=1)
            prog['mini_projects']=fake.random_int(min=1,max=20,step=1)
            prog['certifications_earned']=fake.random_int(min=1,max=10,step=1)
            prog['latest_project_score']=fake.random_int(min=1,max=100,step=1)
            self.prog_list.append(prog)
            df=pd.DataFrame(self.prog_list)
            df= df.drop_duplicates(['programming_id'],keep='first')
            df= df.drop_duplicates(['student_id'],keep='first')
        return df

# creating the dataset for soft skills table and storing it in a excel file
# libraies used : Faker, Pandas
# datasets used from Python : List[], dictionary {}
# Methodology used : Functions, for loop

    def soft_skills(self,ss):
        self.ss_list=[]
        for i in range(1,ss):
            soft_skill={}
            soft_skill['soft_skill_id']=fake.unique.random_int(min=1,max=10000)
            soft_skill['student_id']=i
            soft_skill['communication']=fake.random_int(min=1,max=100,step=1)
            soft_skill['teamwork']=fake.random_int(min=1,max=100,step=1)
            soft_skill['presentation']=fake.random_int(min=1,max=100,step=1)
            soft_skill['leadership']=fake.random_int(min=1,max=100,step=1)
            soft_skill['critical_thinking']=fake.random_int(min=1,max=100,step=1)
            soft_skill['interpersonal_skills']=fake.random_int(min=1,max=100,step=1)
            self.ss_list.append(soft_skill)
            df=pd.DataFrame(self.ss_list)
            df= df.drop_duplicates(['soft_skill_id'],keep='first')
            df= df.drop_duplicates(['student_id'],keep='first')
        return df

# creating the dataset for placements table and storing it in a excel file
# libraies used : Faker, Pandas
# datasets used from Python : List[], dictionary {}
# Methodology used : Functions, for loop

    def placements(self,plt):
        self.placement_list=[]
        for i in range(1,plt):
            plcmnt={}
            plcmnt['placement_id']=fake.unique.random_int(min=1,max=10000)
            plcmnt['student_id']=i
            plcmnt['mock_interview_score']=fake.random_int(min=1,max=100,step=1)
            plcmnt['internships_completed']=fake.random_int(min=1,max=10,step=1)
            plcmnt['placement_status']=fake.random_element(elements=("Ready","Not Ready","Placed","Attending interviews","Awaiting offer","In Final round"))
            if (plcmnt['placement_status'] in ["Placed","Awaiting offer"]):
                plcmnt['company_name']=fake.company()
                plcmnt['placement_package']=fake.random_int(min=700000,max=10000000,step=50000)
            else:
                plcmnt['company_name'] = None
                plcmnt['placement_package']=None
            plcmnt['interview_rounds_cleared']=fake.random_int(min=1,max=5,step=1)
            if (plcmnt['placement_status'] in ["Placed","Awaiting offer"]):
                plcmnt['placement_date']=fake.date_between(start_date="-1y",end_date="+1w")
            else:
                plcmnt['placement_date'] = None
            self.placement_list.append(plcmnt)
            df=pd.DataFrame(self.placement_list)
            df= df.drop_duplicates(['placement_id'],keep='first')
            df= df.drop_duplicates(['student_id'],keep='first')
        return df

# calling the class dataset_creation()
dc_data = dataset_creation()
# calling the function students() using object dc_data and storing it in a excel file
stud_data = dc_data.students(500)
stud_data.to_excel('student.xlsx',index=False)

# calling the function programming() using object dc_data and storing it in a excel file
prog_data = dc_data.programming(500)
prog_data.to_excel('prog.xlsx',index=False)

# calling the function soft_skills() using object dc_data and storing it in a excel file
ss_data = dc_data.soft_skills(500)
ss_data.to_excel('softskills.xlsx',index=False)

# calling the function placements() using object dc_data and storing it in a excel file
placement_data = dc_data.placements(500)
placement_data.to_excel('placement.xlsx',index=False)
```

```python
# Data Storage
# creating database : 
# Methodology used -> Class and methods
# Libraries used -> Pandas, mysql connector, numpy
# importing pandas to access the excel file created using faker
# establishing connection between MYSQL and Python to create database, tables and load the dataset
# usin try...except blocks to identify the connectivity, file opening and other errors

import pandas as pd
import mysql.connector
import numpy as np

# creating a database called plcmnt to store the data in the form of tables
class data_storage:
    def __init__(self,host,user,password):
        self.connection=mysql.connector.connect(
            host="localhost",
            user="root",
            password="893107"
            )
        self.cursor=self.connection.cursor()
    
    def database_creation(self):
        try:
            query="create database if not exists plcmnt"
            self.cursor.execute(query)
            return True
        except mysql.connector.Error as err:
            print(f"Error in database creation:{err}")
            return False
        
# query to use the database created
    def use_db(self):
        try:
            query="use plcmnt"
            self.cursor.execute(query)
            return True
        except mysql.connector.Error as err:
            print(f"Error!!!Unable to use Database plcmnt: {err}")
            return False
        
# create Students table 
    def students_db(self,file):
        try:
            query = """create table if not exists Students(student_id int primary key,
                                               name varchar(100),
                                               age int,
                                               gender varchar(100),
                                               email varchar(100),
                                               phone varchar(20),
                                               enrollment_year int,
                                               course_batch varchar(100),
                                               city varchar(100),
                                               graduation_year int
                                               )"""
            self.cursor.execute(query)
            df = pd.read_excel(file)
            df = df.dropna()
            df['phone']=[data[0]for data in df['phone'].str.split("x")]
            data = [tuple(x) for x in df.values ]
            df=pd.DataFrame(data,columns=['student_id',
                                         'name',
                                         'age',
                                         'gender',
                                         'email',
                                         'phone',
                                         'enrollment_year',
                                         'course_batch',
                                         'city',
                                         'graduation_year'])
            query = """insert into Students(student_id,
                                             name,
                                             age,
                                             gender,
                                             email,
                                             phone,
                                             enrollment_year ,
                                             course_batch,
                                             city,
                                             graduation_year)
                        values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
            for index,row in df.iterrows():
                self.cursor.execute(query,(row['student_id'],row['name'],row['age'],row['gender'],
                              row['email'],row['phone'],row['enrollment_year'],
                              row['course_batch'],row['city'],row['graduation_year']))
            return True
        except mysql.connector.Error as err:
            print(f"Error!!!Unable to load student db: {err}")
            return False
        except FileNotFoundError:
            print(f"Error!!!File not found: {file}")
            return False
        except Exception as err:
            print(f"Error Unexpected: {err}")
            return False
            
# Create Programming table
    def programming_db(self,file):
        try:
            query = """create table if not exists Programming(programming_id int primary key,
                                                student_id int,
                                               language varchar(100),
                                               problems_solved int,
                                               assessments_completed int,
                                               mini_projects int,
                                               certifications_earned int,
                                               latest_project_score int,
                                               foreign key(student_id) references Students(student_id)
                                               )"""
            self.cursor.execute(query)
            df = pd.read_excel(file)
            df = df.dropna()
            data = [tuple(x) for x in df.values ]
            df=pd.DataFrame(data,columns=['programming_id',
                                          'student_id',
                                          'language',
                                          'problems_solved',
                                          'assessments_completed',
                                          'mini_projects',
                                          'certifications_earned',
                                          'latest_project_score'])
            query = """insert into Programming (programming_id,
                                               student_id,
                                               language,
                                               problems_solved,
                                               assessments_completed,
                                               mini_projects,
                                               certifications_earned,
                                               latest_project_score)
                                              values(%s,%s,%s,%s,%s,%s,%s,%s)"""
            for index,row in df.iterrows():
                self.cursor.execute(query,(row['programming_id'],row['student_id'],row['language'],
                          row['problems_solved'],row['assessments_completed'],row['mini_projects'],
                          row['certifications_earned'],row['latest_project_score']))
            return True
        except mysql.connector.Error as err:
            print(f"Error!!!Unable to load programming db: {err}")
            return False
        except FileNotFoundError:
            print(f"Error!!!File not found: {file}")
            return False
        except Exception as err:
            print(f"Error Unexpected: {err}")
            return False
        
# Create Soft_Skills table
    def soft_skills_db(self,file):
        try:
            query = """create table if not exists Soft_Skills(soft_skill_id int primary key,
                                                  student_id int,
                                                  communication int,
                                                  teamwork int,
                                                  presentation int,
                                                  leadership int,
                                                  critical_thinking int,
                                                  interpersonal_skills int,
                                                  foreign key(student_id) references Students(student_id)
                                                  )"""
            self.cursor.execute(query)
            df = pd.read_excel(file)
            df = df.dropna()
            data = [tuple(x) for x in df.values ]
            df=pd.DataFrame(data,columns=['soft_skill_id',
                                          'student_id', 
                                          'communication',
                                          'teamwork',
                                          'presentation',
                                          'leadership',
                                          'critical_thinking',
                                          'interpersonal_skills'])
            query = """insert into Soft_Skills(soft_skill_id,
                                              student_id, 
                                              communication,
                                              teamwork,
                                              presentation,
                                              leadership,
                                              critical_thinking,
                                              interpersonal_skills)
                                              values(%s,%s,%s,%s,%s,%s,%s,%s)"""
            for index,row in df.iterrows():
                row_values = tuple(map(lambda val: int (val) if isinstance(val,(np.int64,np.int32)) else val,row))
                self.cursor.execute(query,row_values)
            return True
        except mysql.connector.Error as err:
            print(f"Error!!!Unable to load soft_skills db: {err}")
            return False
        except FileNotFoundError:
            print(f"Error!!!File not found: {file}")
            return False
        except Exception as err:
            print(f"Error Unexpected: {err}")
            return False
        

# Create Placements table
    def placements_db(self,file):
        try:
            query = """create table if not exists Placements(placement_id int primary key,
                                                student_id int,
                                               mock_interview_score int,
                                               internships_completed int,
                                               placement_status varchar(100),
                                               company_name varchar(100),
                                               placement_package int,
                                               interview_rounds_cleared int,
                                               placement_date varchar(100),
                                               foreign key(student_id) references Students(student_id)
                                               )"""
            self.cursor.execute(query)
            df = pd.read_excel(file)
            df['company_name'] = df['company_name'].fillna('yet to be placed')
            df['placement_package'] = df['placement_package'].fillna(0)
            df['placement_date'] = df['placement_date'].fillna('Not Applicable')
            data = [tuple(x) for x in df.values ]
            df=pd.DataFrame(data,columns=['placement_id',
                                          'student_id',
                                          'mock_interview_score',
                                          'internships_completed',
                                          'placement_status',
                                          'company_name',
                                          'placement_package',
                                          'interview_rounds_cleared',
                                          'placement_date'])
            query = """insert into Placements (placement_id,
                                              student_id,
                                              mock_interview_score,
                                              internships_completed,
                                              placement_status,
                                              company_name,
                                              placement_package,
                                              interview_rounds_cleared,
                                              placement_date)
                                              values(%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
            for index,row in df.iterrows():
                self.cursor.execute(query,(row['placement_id'],row['student_id'],row['mock_interview_score'],
                          row['internships_completed'],row['placement_status'],row['company_name'],
                          row['placement_package'],row['interview_rounds_cleared'],row['placement_date']))
            return True
        except mysql.connector.Error as err:
            print(f"Error!!!Unable to load placements db: {err}")
            return False
        except FileNotFoundError:
            print(f"Error!!!File not found: {file}")
            return False
        except Exception as err:
            print(f"Error Unexpected: {err}")
            return False

#close the database   
    def close_db(self):
        self.connection.close()

# calling the class
# calling the class data_storage()
db_data = data_storage("localhost","root","893107")

# calling the function database_creation() using object db_data
db_creation = db_data.database_creation()
db_use = db_data.use_db()

# calling the function student_db() using object db_data
db_stud = db_data.students_db(r"C:\Users\sathy\OneDrive\Desktop\Project\student.xlsx")

# calling the function programming_db() using object db_data
db_prog = db_data.programming_db(r"C:\Users\sathy\OneDrive\Desktop\Project\prog.xlsx")

# calling the function soft_skills_db() using object db_data
db_ss = db_data.soft_skills_db(r"C:\Users\sathy\OneDrive\Desktop\Project\softskills.xlsx")

# calling the function placements_db() using object db_data
db_placement = db_data.placements_db(r"C:\Users\sathy\OneDrive\Desktop\Project\placement.xlsx")

#to save the data and close the connection using the method close_db()
db_data.connection.commit()
db_data.close_db()
```
